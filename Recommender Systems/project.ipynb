```python
import pandas as pd
import numpy as np
```

<span style='color:Orange'>khandan file ha va tabdil be dataframe</span>

```python
movies=pd.read_csv('movies.csv')
ratings=pd.read_csv('ratings.csv')
movies.head()
```

<span style='color:Orange'>ba setefade az regix sal ra az sotone 'title' estekhrag karde va yek soton be name 'year' egad mikonim.</span>

```python
movies['year']=movies['title'].str.extract('(\(\d\d\d\d\))')
movies['year']=movies['year'].str.extract('(\d\d\d\d)')
movies['title']=movies['title'].str.replace('(\(\d\d\d\d\))','')
movies['title']=movies['title'].str.strip()
movies.head()
```

<span style='color:Orange'>sotoone 'genres' ra ba | split karde va list haye be vogod amade ra jaygozine reshte haye gabli mikonim.</span>

```python
movies['genres']=movies['genres'].str.split('|')
movies.head()
```

<span style='color:Orange'>be ezaye har yek az genre haye mojod yek sotoon misazim. megdar har genre braye yek movie agar dar an genre bashad 1, dar geyre in soorat 0 khahad bood.</span>

```python
moviesWithGenres=movies.copy()
for index, row in movies.iterrows():
    for genre in row['genres']:
        moviesWithGenres.at[index, genre]=1
moviesWithGenres.fillna(0, inplace=True)
moviesWithGenres.head()
```

```python
ratings.head()
```

<span style='color:Orange'>sotoone 'timestamp' ra hazf mikonim chon niyazi be an nadarim.</span>

```python
ratings=ratings.drop('timestamp', 1)
```

```python
ratings.head()
```

<span style='color:Orange'>yek dataframe shamle 'title' va 'rating' misazim ke farz mishavad marboot be karbari ast ke mikhahim film pishnahad bedahim.</span>

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

## 1 - Content-Based recommendation system

<span style='color:Orange'>az dataframe movies satr haye movie haie ke karbar dide ra peyda karde va ba inputMovies edgam mikonim ta 'movieId' ra dashte bashim amma 'genres' va 'year' ra hazf mikonim chon be anha niyaz nadarim.</span>

```python
inputId=movies[movies.title.isin(inputMovies.title)]
inputMovies=pd.merge(inputId, inputMovies)
inputMovies=inputMovies.drop('genres',1).drop('year',1)
inputMovies
```

<span style='color:Orange'>az dataframe 'moviesWithGenres' genre haye movie haye karbar ra mikhanim va baz ham sotoon haye ezafe ra hazf mikonim.</span>

```python
inputGenres=moviesWithGenres[moviesWithGenres.title.isin(inputMovies.title.tolist())]
inputGenres=inputGenres.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
inputGenres.reset_index(drop=True, inplace=True)
inputGenres
```

<span style='color:Orange'>ba zarbe (zarbe matrisi) matris ke dar cell gable be das amad da sotoon inputMovies['rating'] matrise egad mishavad ke an ra 'userProfile' minamin, in matris be noie mizan alaghemandi karbar be har yek az genre ha, ba tavajoh be genre film haye ke dide va emtiyazi ke be an ha dade moshakhas mikonad. harche adad be dast amade braye yek genre bishtar bashad yani alaghemandi karbar be an bishtar ast.</span>

```python
userProfile=inputGenres.transpose().dot(inputMovies['rating'])
userProfile
```

<span style='color:Orange'>az dataframe 'moviesWithGenres' sotoon haye genre ha ra bardashte va dar yek dataframe digar be name 'genreTable' garar midahim. 'movieId' ra niz be onvane index in dataframe set mikonim.</span>

```python
genreTable=moviesWithGenres.drop('title',1).drop('genres',1).drop('year',1)
genreTable=genreTable.set_index('movieId')
genreTable.head()
```

<span style='color:Orange'>userProfile ra dar genreTable zarbe nazir be nazir mikonim va magadir hasel baraye har film ra jam mikonim va sepas be majmoe userProfile tagsim mikonim. be bayane digar miyangim vazn var migirm. in be ma neshan midahad har film ba tavajoh be genre hayi ke darad va alaghemandi karbar be in genre ha ke az roye film hayi ke gablal dide hads zadim che gadr khob ast, yani karbar chegad ehtemalan an ha ra mipasandad. harche addad be dast amade braye yek film bishtar bashad yani an film shanse bishtari braye entekhab tavasote karbar darad.</span>

```python
recommendationTable=((userProfile*genreTable).sum(axis=1))/userProfile.sum()
recommendationTable.head()
```

<span style='color:Orange'>recommendationTable be tartibe nozooli morattab mikonim.</span>

```python
recommendationTable=recommendationTable.sort_values(ascending=False)
recommendationTable.head()
```

<span style='color:Orange'>20 film ba balatarin shans ra namayesh midahim.</span>

```python
movies[movies.movieId.isin(recommendationTable.head(20).keys())]
```

## 2 - Collaborative Filtering

### 2-1 - User-based

<span style='color:Orange'>user hayi ke hade agal yek film az film hayi ke karbar morede nazare ma dide ast ra dide and, joda mikonim.</span>

```python
userSubset=ratings[ratings.movieId.isin(inputMovies.movieId)]
userSubset.head()
```

<span style='color:Orange'>userSubset ra bar asase 'userId' grouh bandi mikonim va dar dataframe jadidi be name 'userSubsetGroup' garar midahim.</span>

```python
userSubsetGroup=userSubset.groupby('userId')
userSubsetGroup.head()
```

<span style='color:Orange'>userSubsetGroups ra bar asase inke har user chan filme moshtarak darad be sorat nozooli moratab mikonim.</span>

```python
userSubsetGroup=sorted(userSubsetGroup, key=lambda x:len(x[1]), reverse=True)
userSubsetGroup[:3]
```

<span style='color:Orange'>baraye inke mohasebat ziyad tool nakeshad 100 karbar aval ra entekhab mikonim.</span>

```python
userSubsetGroup=userSubsetGroup[:100]
```

<span style='color:Orange'>ba stefade az pearson correlation mizan shebahat har yek a 100 karbar entkhabi ba karbar mored nazar ra mohasebe mikonim.</span>

```python
from math import sqrt
pearsonDict={}
for name, group in userSubsetGroup:
    # group va inputMovies ra bar asase 'movieId' moratab mikonim to motmaen bashim tartib anha dar har do yek joor ast.
    group=group.sort_values(by='movieId')
    inputMovies=inputMovies.sort_values(by='movieId')
    # len(group) dar formol pearson correlation neyiaz ast.
    nRatings=len(group)
    # id movie hayi ke karbare digar(!= karbar hadaf) dide ra braye karbar hadaf peyda karde va joda mikonim.
    tempDf=inputMovies[inputMovies['movieId'].isin(group['movieId'].tolist())]
    # list emtiyazati ke in do karbar be film haye moshtarak dade and ra tashkil midahim.
    tempUserRating=tempDf['rating'].tolist()
    tempGroupRating=group['rating'].tolist()
    # az inja be bad mohasebeye pearson correlation va grar dadane hasel an dar dictionary ast.
    Sxx = sum([i**2 for i in tempUserRating]) - pow(sum(tempUserRating),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupRating]) - pow(sum(tempGroupRating),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempUserRating, tempGroupRating)) - sum(tempUserRating)*sum(tempGroupRating)/float(nRatings)
    if Sxx!=0 and Syy!=0:
        pearsonDict[name]=Sxy/sqrt(Sxx*Syy)
    else:
        pearsonDict[name]=0
    
```

```python
pearsonDict
```

<span style='color:Orange'>dictionary ra be dataframe tabdil mikonim</span>

```python
pearsonDf=pd.DataFrame.from_dict(pearsonDict, orient='index')
pearsonDf.columns=['similarityIndex']
pearsonDf['userId']=pearsonDf.index
pearsonDf=pearsonDf.reset_index(drop=True)
pearsonDf.head()
```

<span style='color:Orange'>pearsonDf ra bar asase 'similarityIndex' be sorat nozooli moratab mikonim va 50 satre aval ra bar midarim</span>

```python
topUsers=pearsonDf.sort_values(by='similarityIndex', ascending=False)[:50]
topUsers.head()
```

<span style='color:Orange'>topUsers ra ba dataframe ratings edgam mikonim ta emtiyazi ke har yek az in afrad be har yek az film hayi ke dide ast be dast biyavarim.</span>

```python
topUsersRating=topUsers.merge(ratings)
topUsersRating.head()
```

<span style='color:Orange'>ba zarbe 'similarityIndex' dar 'rating', emtiyaze vazn dar ra be dast miavarim.</span>

```python
topUsersRating['weightedRating']=topUsersRating['similarityIndex']*topUsersRating['rating']
topUsersRating.head()
```

<span style='color:Orange'>topUsersRating ra bar asase 'movieId' grooh bandi karde va roye har do sotoon sum mizanim.</span>

```python
topUsersRating=topUsersRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
topUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
topUsersRating.head()
```

<span style='color:Orange'>sotoon 2 ra be sotoon 1 tagsim mikonim ta miyangin vazn dar emtiyaz be har film ra be dast avarim.</span>

```python
recommendationTable=pd.DataFrame()
recommendationTable['weighted average recommendation score']=topUsersRating['sum_weightedRating']/topUsersRating['sum_similarityIndex']
recommendationTable['movieId'] = topUsersRating.index
recommendationTable.head()
```

<span style='color:Orange'>film ha ra bar asase 'weighted average recommendation score' be sorate nozooli moratab mikonim. film haye ba emtiyaze balatar shanse bishtary baraye entkhab tavasot karbar darand chon be tore miyangin emtiyaze balatari az afradi ke shebahat ziyadi be in karbar darand grefte and.</span>

```python
recommendationTable=recommendationTable.sort_values(by='weighted average recommendation score', ascending=False)
recommendationTable.head()
```

### 2-2 - Item-based

<span style='color:Orange'>ba estefade az pivot dataframe ratings ra be sorat zir tagiere shekl midahim va magadir NaN ra ba 0 por mikonim.</span>

```python
moviesUsers=pd.pivot(ratings, columns='userId', index='movieId', values='rating')
moviesUsers=moviesUsers.fillna(0)
moviesUsers.head()
```

<span style='color:Orange'>ba estefade az NearestNeighbors 3 (dar in ja) nazdik tarin film be har film ra peyda mikonim.</span>

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(moviesUsers.values)
distances, indices = knn.kneighbors(moviesUsers.values, n_neighbors=3)
```

<span style='color:Orange'>masalan 3 nazdiktarin film be film aval be tartib: khodash, filme 2354 (ba shoro az 1) va film 419 mibashand.</span>

```python
indices
```

<span style='color:Orange'>dar in array niz fasele anha garar darad.</span>

```python
distances
```

<span style='color:Orange'>dar cell zir emtiyaze har yek az film hayi ke karbar nadide ra batavajoh be shebahat be film hayi ke dide ast takhmin mizanim. braye in kar az formole zir estefade mishavad :</span>
    
R(m, u) = {âˆ‘ â±¼ S(m, j)R(j, u)}/ âˆ‘ â±¼ S(m, j)
- R(m, u): the rating for movie m by user u
- S(m, j): the similarity between movie m and movie j. S(m, j)=1-distances(m,j)
- j âˆˆ J where J is the set of the similar movies to movie m

```python
recommendationTable=pd.DataFrame(columns=['movieId', 'predicted_r'])
number_neighbors=3
for r, i in list(enumerate(moviesUsers.index)):
    if i not in inputMovies.movieId.tolist():
        sim_movies = indices[r].tolist()
        movie_distances = distances[r].tolist()
        if r in sim_movies:
            id_movie = sim_movies.index(r)
            sim_movies.remove(r)
            movie_distances.pop(id_movie)
        else:
            sim_movies = sim_movies[:number_neighbors-1]
            movie_distances = movie_distances[:number_neighbors-1]
        movie_similarity = [1-x for x in movie_distances]
        movie_similarity_copy = movie_similarity.copy()
        nominator = 0
        for s in range(0,len(movie_similarity)):
            mi=moviesUsers.index.tolist()[sim_movies[s]]
            if mi not in inputMovies.movieId.tolist():
                if len(movie_similarity_copy) == (number_neighbors - 1):
                    movie_similarity_copy.pop(s)
                else:
                    movie_similarity_copy.pop(s-(len(movie_similarity)-len(movie_similarity_copy)))
            else:
                nominator = nominator + movie_similarity[s]*float(inputMovies[inputMovies['movieId']==mi]['rating'])
        if len(movie_similarity_copy)>0:
            if sum(movie_similarity_copy) > 0:
                recommendationTable.loc[len(recommendationTable)]=[i, nominator/sum(movie_similarity_copy)]
            else:
                recommendationTable.loc[len(recommendationTable)]=[i, 0]
        else:
            recommendationTable.loc[len(recommendationTable)]=[i, 0]
            
    
```

<span style='color:Orange'>bar asase 'predicted_r' be sorat nozooli moratab mikonim.</span>

```python
recommendationTable=recommendationTable.sort_values(by='predicted_r', ascending=False)
recommendationTable.head(10)
```
