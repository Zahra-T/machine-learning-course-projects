```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('housePrice.csv', na_values=' ')
```

```python
# drop samples with unknown addresses
df.dropna(axis=0, how='any', inplace=True)
```

```python
for item in df['Area']:
    if ',' in item:
        df['Area'].replace({item:item.replace(',','').replace(' ','')}, inplace=True)
df['Area']=pd.to_numeric(df['Area'])
```

```python
# drop outliers
df=df[df['Area']<=2000]
```

```python
x=df[['Area', 'Room', 'Parking', 'Warehouse', 'Elevator', 'Address']]
y=df['Price(USD)']
```

```python
x=pd.get_dummies(data=x, columns=['Parking', 'Warehouse', 'Elevator', 'Address'])
x
```

```python
msk=np.random.rand(len(df))<0.8
train_x=x[msk]
train_y=y[msk]
test_x=x[~msk]
test_y=y[~msk]
```

```python
from sklearn import linear_model
```

```python
regr=linear_model.LinearRegression()
regr.fit(train_x, train_y)
print('Coefficients: ', regr.coef_)
print("Intercept: ", regr.intercept_)
```

```python
from sklearn.metrics import r2_score
```

```python
test_y=np.asanyarray(test_y)
test_y_=regr.predict(test_x)
print("Mean absolute error: %.2f"%np.mean(np.absolute(test_y_-test_y)))
print('Residual sum of squares (MSE): %.2f'%np.mean((test_y_-test_y)**2))
print("R2_score: %.2f"%r2_score(test_y, test_y_))
```
